```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import MinMaxScaler
import math

import keras
from keras.models import Sequential
from keras.layers import Dense,Dropout,LSTM
```

```python
df= pd.read_excel('../../DATA/bitcoin.xlsx')
```

```python
df
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df['date']= pd.to_datetime(df[['year','month','day']])
```

```python
df
```

```python
df.drop(['year','month','day','Stock Splits', 'Dividends'], axis=1,inplace=True)
```

```python
df.set_index('date',inplace=True)
```

```python
df
```

```python
data= df[['Close']].copy()
```

```python
data=data.values
```

```python
data
```

```python
scaler=MinMaxScaler(feature_range=(0,1))
```

```python
scaled_data= scaler.fit_transform(data)
```

```python
scaled_data
```

```python
training_data= math.ceil(len(scaled_data)*0.8)
```

```python
training_data
```

```python
train_data=scaled_data[0:training_data,:]
```

```python
train_data
```

```python
prediction_days=60

x_train=[]
y_train=[]

for x in range(prediction_days,len(train_data)):
    x_train.append(train_data[x-prediction_days:x,0])
    y_train.append(train_data[x,0])
```

```python
x_train,y_train= np.array(x_train), np.array(y_train)
```

```python

x_train= np.reshape(x_train,(x_train.shape[0],x_train.shape[1],1))
```

```python
model= Sequential()
model.add(LSTM(50, return_sequences=True,input_shape=(x_train.shape[1],1)))
model.add(Dropout(0.2))
model.add(LSTM(50,return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(50))
model.add(Dropout(0.2))
model.add(Dense(units=25))
model.add(Dropout(0.2))
model.add(Dense(1))
```

```python
model.compile(optimizer='adam', loss='mean_squared_error',metrics= 'mean_absolute_error')
```

```python
history=model.fit(x_train,y_train, epochs=1000, batch_size=64)
```

```python
model.summary()
```

```python
test_data= scaled_data[training_data-60:,:]
```

```python
x_test=[]
y_test=data[training_data:,:]
for i in range (60, len(test_data)):
    x_test.append(test_data[i-60:i,0])
```
