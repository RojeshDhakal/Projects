Date cannot be stored in excel straight away therefore we need to convert it into year, month and day and then we can save the file in excel. 

```python
btc_hist['year']= pd.to_datetime(btc_hist['Date']).dt.year
```

```python
btc_hist
```

```python
btc_hist['month']= pd.to_datetime(btc_hist['Date']).dt.month
```

```python
btc_hist
```

```python
btc_hist['day']= pd.to_datetime(btc_hist['Date']).dt.day
```

```python
btc_hist
```

```python
btc_hist.drop('Date',axis=1,inplace=True)
```
