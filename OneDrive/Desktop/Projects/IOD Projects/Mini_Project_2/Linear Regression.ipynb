## Choosing the feature

```python
X= data.drop(['Close','Volume'],axis=1)
```

## Train/Test Split

Since our data is time series data we cannot split it ramdomly using scikit learn so we need to random in a order so that it have a trend.

```python
number=500
```

```python
X_train=X[:number]
```

```python
X_test=X[number:]
```

```python
y_train=y[:number]
```

```python
y_test=y[number:]
```

```python
model=LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
preds=model.predict(X_test)
```

```python
model.score(X_test,y_test)
```

```python
mean_squared_error(y_test, preds, squared=True)
```

```python
r2 = r2_score(y_test, preds)
print(f"R-squared: ", r2)
```

```python
preds.shape
```

```python
predictions=pd.DataFrame(preds,index=y_test.index)
predictions
```

```python
fig,ax=plt.subplots(figsize=(12,8), facecolor='green')
ax.plot(y_test, color='blue')
ax.plot(predictions, color='green')
ax.legend(['Actual','Predicted'])
ax.set_title('Actual Vs Predicted')
ax.set_xlabel('Date')
ax.set_ylabel('Closing Price')
plt.show()

fig.savefig('ActualVsPredicted.png')
```
