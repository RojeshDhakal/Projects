```python
import numpy as np
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_excel('../DATA/AAPLStocks.xlsx')
data.drop('Unnamed: 0', axis=1, inplace=True)
data['date']= pd.to_datetime(data[['Year', 'month','day']])
data.set_index('date', inplace=True)
data.drop(['Year', 'month','day'], axis=1,inplace=True)
data
```

```python
df=data[['Close']].copy()
df
```

```python
df.plot()
```

```python
df.describe()
```

```python
from statsmodels.tsa.stattools import adfuller
result= adfuller(df.Close.dropna())
print(f"ADF Statistic: {result[0]}")
print(f"p-value: {result[1]}")
```

p value is more accept  null hypothesis which means data is not stationary.
 

if the data is not stationary then we need to make it stationary. By differencing


### Differencing

means we shift one position of the data

```python
df['close_differencing']= df['Close'] - df['Close'].shift(-1)
```

```python
df['Close'].shift(1)
```

```python
df
```

```python
result= adfuller(df.close_differencing.dropna())
print(f"ADF Statistic: {result[0]}")
print(f"p-value: {result[1]}")
```

Reject null hypothesis and accepts alternate which proves that data is stationary.

```python
df.close_differencing.plot() 
```

```python
from statsmodels.graphics.tsaplots import plot_acf
```

```python
fig,(ax1,ax2)= plt.subplots(1,2, figsize=(16,8))
ax1.plot(df.close_differencing.dropna())
ax1.set_title('Difference Once')
plot_acf(df.close_differencing.dropna(),ax=ax2);
```

sudden decrease or shut off which is Ar value (p=1)

```python
from statsmodels.graphics.tsaplots import plot_pacf
```

```python
fig,(ax1,ax2)= plt.subplots(1,2, figsize=(16,20))
ax1.plot(df.close_differencing.dropna())
ax1.set_title('Difference Once')
ax2.set_ylim(0,1)
plot_pacf(df.close_differencing.dropna(),ax=ax2);
```

q might be 3 beacuse it has exponential decrease but it has decrease to 1

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
model= ARIMA(df.Close, order=(1,1,3))
model_fit= model.fit()
```

```python
model_fit.summary()
```

```python
df['forecast']= model_fit.predict(start=1000, end= 5000)
df[['Close', 'forecast']].plot()
```

```python
import statsmodels.api as sm
```

```python
model= sm.tsa.statespace.SARIMAX(df['Close'], order=(1,1,3), seasonal_order=(1,1,3,365))
results= model.fit()
```

```python
df['new_forecast']=results.predict(start=1000,end=5000,dynamic=True)
df[['Close','new_forecast']].plot()
```
