```python
import math

from scipy import stats
```

```python
t2, p2 = stats.ttest_ind(Nz2.inflation, Nz2.interest_rate)
print("t = " + str(t2))
print("p = " + str(p2))
```

```python
def print_sig(p_value, alpha):
    if p_value < alpha:
        print("We reject our null hypothesis.")
    elif p_value > alpha:
        print("We fail to reject our null hypothesis.")
    else:
        print("Our test is inconclusive.")
```

```python
alpha=0.05
```

```python
print_sig(p2, alpha)
```

```python
t2, p2 = stats.ttest_ind(Nz2.inflation, Nz2.unemployment)
print("t = " + str(t2))
print("p = " + str(p2))
print_sig(p2, alpha)
```
