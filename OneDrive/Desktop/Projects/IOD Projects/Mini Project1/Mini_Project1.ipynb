```python
import math

from scipy import stats
```

```python
alpha=0.05
```

```python
def print_sig(p_value, alpha):
    if p_value < alpha:
        print("We reject our null hypothesis.")
    elif p_value > alpha:
        print("We fail to reject our null hypothesis.")
    else:
        print("Our test is inconclusive.")
```

# Hypothesis test between inflation and interest_rate

```python
t2, p2 = stats.ttest_ind(nz.inflation,nz.interest_rate)
                    
print("t = " + str(t2))
print("p = " + str(p2))
print_sig(p2, alpha)
```

# Hypothesis test between inflation and unemployment

```python
t2, p2 = stats.ttest_ind(nz.inflation,nz.unemployment)
                    
print("t = " + str(t2))
print("p = " + str(p2))
print_sig(p2, alpha)
```
